```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom

num_validators = [2**14,2**15,2**16,2**17,2**18,2**19,2**20]
num_subnets = 64
PERSISTENT_COMMITTEE_PERIOD = 2**11
avg_validators_per_subnet = [v/num_subnets for v in num_validators]
avg_validators_per_subnet



```

```python
n = num_validators[0]
p = 1/num_subnets
k = range(int(avg_validators_per_subnet[0]*4/5),int(avg_validators_per_subnet[0]*6/5))

fig, ax = plt.subplots(figsize=(14,7))
ax.plot(k, binom.pmf(k, n, p), 'ro')
ax.vlines(k, 0, binom.pmf(k, n, p), colors='r', lw=1, alpha=0.5)
ax.grid(b=True, which='both', axis='both')
plt.title('probability that a particular subnet is selected N times with 2^14 validators')
plt.xticks(np.arange(k[0], k[len(k)-1], step=5))
plt.yticks(np.arange(0, .03, step=0.005))
plt.show()
```

```python
n = num_validators[0]
p = 1/num_subnets
k = range(int(avg_validators_per_subnet[0]*.9),int(avg_validators_per_subnet[0]*1.1))
fig, ax = plt.subplots(figsize=(14,4))
ax.plot(k, binom.cdf(k, n, p), 'ro')
ax.vlines(k, 0, binom.cdf(k, n, p), colors='r', lw=1, alpha=0.5)
ax.grid(b=True, which='both', axis='both')
plt.title('probability that a particular subnet is selected <= N times with 2^14 validators')
plt.xticks(np.arange(k[0], k[len(k)-1], step=5))
plt.yticks(np.arange(0, 1.1, step=0.1))
plt.show()
```
