```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom

num_validators = [2**14,2**15,2**16,2**17,2**18,2**19]
num_subnets = 64
avg_validators_per_subnet = [v/num_subnets for v in num_validators]
avg_validators_per_subnet
```

```python
n = num_validators[0]
p = 1/num_subnets
k = range(int(avg_validators_per_subnet[0]*4/5),int(avg_validators_per_subnet[0]*6/5))

fig, ax = plt.subplots(figsize=(14,7))
ax.plot(k, 1-binom.pmf(k, n, p), 'ro')
ax.vlines(k, 0.96, 1-binom.pmf(k, n, p), colors='r', lw=1, alpha=0.5)
ax.grid(b=True, which='both', axis='both')
plt.title('probability that a particular subnet is selected N times with 2^14 validators')
plt.xticks(np.arange(k[0], k[len(k)-1], step=5))
plt.yticks(np.arange(.96, 1.0, step=0.005))
plt.show()

n = num_validators[0]
p = 1/num_subnets
k = range(220,280)
fig, ax = plt.subplots(figsize=(14,7))
ax.plot(k, 1-binom.cdf(k, n, p), 'ro')
ax.vlines(k, 0, 1-binom.cdf(k, n, p), colors='r', lw=1, alpha=0.5)
ax.grid(b=True, which='both', axis='both')
plt.title('probability that a particular subnet is selected > N times with 2^14 validators')
plt.xticks(np.arange(k[0], k[len(k)-1], step=5))
plt.yticks(np.arange(0, 1.1, step=0.1))
plt.show()

P_240 = sum(binom.pmf(list(range(240)),2**14, 1/64))
P_240
```

```python
num_validating_nodes = [2**9,2**10,2**11,2**12,2**13,2**14]


fig, ax = plt.subplots(figsize=(16,8))

lines = []
labels = []
nv = num_validators[0]
for nvn in num_validating_nodes[:len(num_validating_nodes)]:
    avn = float(nv)/float(nvn)
    n = int(nvn)
    p = min(avn,num_subnets)/num_subnets
    k = range(210,300)
    lines.append(ax.plot(k, 1-binom.cdf(k, n, p), '--'))
    labels.append(str(n)+' vn,'+str(avn)+' avn')

ax.grid(b=True, which='both', axis='both')
ax.legend(labels,loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('probability that a particular subnet is selected > N times with ' + str(nv) + ' validators')
plt.xticks(np.arange(210, 300, step=4))
plt.yticks(np.arange(0, 1.1, step=0.1))
plt.show()

fig, ax = plt.subplots(figsize=(16,8))

lines = []
labels = []
nv = num_validators[1]
for nvn in num_validating_nodes[:len(num_validating_nodes)]:
    avn = float(nv)/float(nvn)
    n = int(nvn)
    p = min(avn,num_subnets)/num_subnets
    k = range(450,550)
    lines.append(ax.plot(k, 1-binom.cdf(k, n, p), '--'))
    labels.append(str(n)+' vn,'+str(avn)+' avn')

ax.grid(b=True, which='both', axis='both')
ax.legend(labels,loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('probability that a particular subnet is selected > N times with ' + str(nv) + ' validators')
plt.xticks(np.arange(450,550, step=4))
plt.yticks(np.arange(0, 1.1, step=0.1))
plt.show()


fig, ax = plt.subplots(figsize=(16,8))

lines = []
labels = []
nv = num_validators[2]
for nvn in num_validating_nodes[1:len(num_validating_nodes)]:
    avn = float(nv)/float(nvn)
    n = int(nvn)
    p = min(avn,num_subnets)/num_subnets
    k = range(940,1080)
    lines.append(ax.plot(k, 1-binom.cdf(k, n, p), '--'))
    labels.append(str(n)+' vn,'+str(avn)+' avn')

ax.grid(b=True, which='both', axis='both')
ax.legend(labels,loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('probability that a particular subnet is selected > N times with ' + str(nv) + ' validators')
plt.xticks(np.arange(940,1080, step=6))
plt.yticks(np.arange(0, 1.1, step=0.1))
plt.show()


```
